# Classification
## 阈值和混淆矩阵
假设您有一个用于垃圾邮件检测的逻辑回归模型，该模型预测 0 到 1 之间的值，表示给定电子邮件是垃圾邮件的概率。预测值为 0.50 表示电子邮件为垃圾邮件的可能性为 50%，预测值为 0.75 表示电子邮件为垃圾邮件的可能性为 75%，依此类推。 
您希望在电子邮件应用程序中部署此模型，以将垃圾邮件过滤到单独的邮件文件夹中。但要做到这一点，您需要将模型的原始数值输出（例如 0.75）转换为两个类别之一：“垃圾邮件”或“非垃圾邮件”。 
要进行此转换，您需要选择一个阈值概率，称为**分类阈值**。然后，概率高于阈值的示例将被分配给正类，即您正在测试的类（此处为垃圾邮件）。概率较低的示例被分配给负类，即替代类（此处不是垃圾邮件）。
### 混淆矩阵
概率分数不是现实，也不是基本事实。二元分类器的每个输出有四种可能的结果。对于垃圾邮件分类器示例，如果将基本事实布置为列，将模型的预测布置为行，则结果如下表（称为混淆矩阵）：
|       |真的真|真的假|阈值改变|
|------|------|-----|------|
|预测的真|TP   |FP   |一起变  |
|预测的假|FN   |TN   |一起变  |

依据混淆矩阵数据分类：
1. 分开式：正例和负例通常属于同一类 差异很好，大多数正例的分数高于 反例。
2. 不分隔：许多正例的得分低于 而许多负例的得分高于 正例。
3. 不平衡：仅包含几个正类别的样本。
## 准确率、召回率、精确率和相关指标


### 准确率
是$\frac{TP+TN}{TP+TN+FN+FP}$
### 召回率或真正例率(TPR)
$$TPR=\frac{TP}{TP+FN}$$
### 假正例率(FPR)
$$FPR=\frac{FP}{FP+TN}$$
### 精确率
$$Precesion = \frac{TP}{TP+FP}$$

## ROC 和曲线下面积AUC
### 接收者操作特征曲线 (ROC)
ROC 曲线的绘制方法是计算真正例率 (TPR) 和假正例率 (FPR) 值
### 曲线下方面积 (AUC)
ROC 曲线下面积 (AUC) 表示模型的概率， 如果给定一个随机选择的正类别和负类别样本， 正值将大于负值。
### 用于选择模型和阈值的 AUC 和 ROC
曲线下面积是一个非常有用的指标，用于比较两个不同模型 只要数据集大致均衡即可。（请参阅精确率与召回率曲线， 针对不平衡的数据集。）下方是较大区域 那么曲线通常效果最佳

## 预测偏差
预测偏差是指一个模型 预测 求平均值， 标准答案标签 数据。使用数据集训练的模型 其中 5% 的电子邮件是垃圾邮件，平均而言， 被归类为垃圾邮件的电子邮件。也就是说， 标准答案数据集的概率为 0.05，那么模型预测的平均值应为 也为 0.05。在这种情况下，模型的预测偏差为零。/ 那么模型可能仍存在其他问题。
如果模型有 50% 的时间预测电子邮件是垃圾邮件，那么 训练数据集存在问题，模型的新数据集 模型本身。不限 二者之间的显著差异表明模型 某些预测偏差。

导致预测偏差的原因可能是：

1. 数据中的偏差或噪声，包括训练集的偏差抽样
2. 正则化过于强烈，意味着模型过度简化， 必要的复杂性
3. 模型训练流水线中的 bug
4. 提供给模型的特征集不足以完成任务
